<!-- Make A Decision Tree with Python module -->

```python
#Three lines to make our compiler able to draw:
import sys
import matplotlib
# matplotlib.use('Agg')
# matplotlib.use( 'tkagg' )

import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pandas.read_csv("Lab1_DT1_data.csv")

d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)

#Two  lines to make our compiler able to draw:
# plt.savefig(sys.stdout.buffer)
plt.show()
# sys.stdout.flush()

#NOTE:
#You will see that the Decision Tree gives you different results if you run it enough times, even if you feed it with the same data.

#That is because the Decision Tree does not give us a 100% certain answer. It is based on the probability of an outcome, and the answer will vary.
```

<!-- DECISION TREE FROM SLIDE'S DATASET -->

```python
#Three lines to make our compiler able to draw:
import sys
import matplotlib
# matplotlib.use('Agg')
# matplotlib.use( 'tkagg' )

import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pandas.read_csv("Lab1_DT2_slide.csv")

#Age, Education, Income, Marital Status, Buy Computer
d={"<18":0,"18-35":1,"36-55":2,">55":3}
df['Age']=df['Age'].map(d)

d={"High School":0,"Masters":1, "Bachelors":2}
df["Education"] = df["Education"].map(d)

d={'Low':0,'High':1}
df['Income']=df['Income'].map(d)

d={'Single':0,'Married':1}
df['Marital Status']=df['Marital Status'].map(d)

d = {'YES': 1, 'NO': 0}
df['Buy Computer'] = df['Buy Computer'].map(d)

features = ['Age', 'Education', 'Income', 'Marital Status']

X = df[features]
y = df['Buy Computer']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)

#Two  lines to make our compiler able to draw:
# plt.savefig(sys.stdout.buffer)
plt.show()
# sys.stdout.flush()

#NOTE:
#You will see that the Decision Tree gives you different results if you run it enough times, even if you feed it with the same data.

#That is because the Decision Tree does not give us a 100% certain answer. It is based on the probability of an outcome, and the answer will vary.
```
